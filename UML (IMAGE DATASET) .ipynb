### <font color='green'>PCA ON MNIST IMAGE DATASET</font>

#### <font color='blue'>PCA (Principal Component Analysis): </font>
Principal Component Analysis (PCA) is a technique used to reduce the dimensionality of the data while preserving the relevant information. 

It works by transforming the data into a set of orthonormal components which explain the variance in the data. 

PCA is typically used in exploratory data analysis tasks, and works best when the variables are highly correlated and there is a large amount of data.

#### <font color='blue'>LDA (Linear Discriminant Analysis):</font>
Linear Discriminant Analysis (LDA) is a supervised machine learning technique used to reduce the dimensionality of the data while preserving the relevant information.

It works by projecting the data onto a set of linear discriminants in order to maximize the separation between different classes of data. LDA is usually used in classification tasks, and works best when there is a smaller amount of data and the classes are well-separated.


 ### <font color='blue'> Import Libraries </font>

```python
import pandas as pd
import numpy as np
import gzip
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
```

 ### <font color='blue'>Import training set </font>

```python
with gzip.open('D:\\train-images-idx3-ubyte.gz', 'rb') as f:
    traindata = np.frombuffer(f.read(), np.uint8, offset=16)
```

 ### <font color='blue'>Import training labels  </font>

```python
with gzip.open('D:\\train-labels-idx1-ubyte.gz', 'rb') as f:
    trainlabels = np.frombuffer(f.read(), np.uint8, offset=8)
```

 ### <font color='blue'>Import testing set  </font>

```python
with gzip.open('D:\\t10k-images-idx3-ubyte.gz', 'rb') as f:
    testdata = np.frombuffer(f.read(), np.uint8, offset=16)
   
```

 ### <font color='blue'>Import testing labels  </font>

```python
with gzip.open('D:\\t10k-labels-idx1-ubyte.gz', 'rb') as f:
    testlabels = np.frombuffer(f.read(), np.uint8, offset=8)
```

### <font color='blue'>Reshape the data and labels</font>

```python
#reshape the datasets
traindata = traindata.reshape(-1, 784)
testdata = testdata.reshape(-1, 784)
```

 ### <font color='blue'>Standardized The Dataset  </font>

```python
#standardizing the data
scaler = StandardScaler()
scaler.fit(traindata)
scaled_traindata = scaler.transform(traindata)
scaled_testdata = scaler.transform(testdata)
```

### <font color='blue'>Applying PCA </font>

```python
#applying PCA
pca = PCA(n_components=2)
pca.fit(scaled_traindata)
pca_traindata = pca.transform(scaled_traindata)
pca_testdata = pca.transform(scaled_testdata)
```

### <font color='blue'>Plotting the Result  </font>

```python
#plotting the results for PCA
plt.scatter(pca_traindata[:, 0], pca_traindata[:, 1], c=trainlabels, cmap=plt.cm.Set1)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('PCA MNIST dataset')
plt.show()
```

### <font color='green'>LDA ON MNIST IMAGE DATASET</font>

```python
#applying LDA
lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(scaled_traindata, trainlabels)
lda_traindata = lda.transform(scaled_traindata)
lda_testdata = lda.transform(scaled_testdata)
```

### <font color='blue'>Plotting the Result <font>

```python
#plotting the results for LDA
plt.scatter(lda_traindata[:, 0], lda_traindata[:, 1], c=trainlabels, cmap=plt.cm.Set1)
plt.xlabel('LDA 1')
plt.ylabel('LDA 2')
plt.title('LDA MNIST dataset')
plt.show()
```

### <font color='green'>OBSERVATIONS </font>

The plots obtained from PCA and LDA showcase the results of the dimensionality reduction techniques on the MNIST dataset. Both of the techniques were able to reduce the dimensionality of the data from 784 to 2, allowing for a better visualization of the data. 

### <font color='green'> GRAPH OBSERVATIONS </font>
Based on the two plots, it is evident that PCA was able to capture the structure of the data while LDA was able to capture the differences between the classes.

PCA is an unsupervised learning technique, meaning it does not take into account any labels associated with the data. This is why the PCA plot shows a relatively disorganized structure of the data points, which are distributed across the two dimensions. 

On the other hand, LDA is a supervised learning technique, meaning it takes into account the labels associated with the data. This is why the LDA plot shows a more organized structure of the data points with each class being clearly distinguishable from the other. 

Overall, PCA and LDA are both efficient dimensionality reduction techniques and can be used to visualize the datasets with high-dimensional features.
