### <font color='BLUE'> SVD - IRIS DATASET </font>

<font color="GREEN">Singular Value Decomposition (SVD) </font>is an algorithm used for dimensionality reduction which is used to reduce the data size while preserving the important features. It is one of the most powerful techniques in machine learning and it is used in many applications such as image and video processing, machine learning, and natural language processing. 

```python
#importing libraries 
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
#loading the iris dataset
iris = load_iris()
#creating a dataframe
X = pd.DataFrame(iris.data, columns = iris.feature_names)
y = pd.DataFrame(iris.target, columns = ['Target'])
data = pd.concat([X,y], axis = 1)
```

```python
#displaying the dataframe 
data
```

```python
data.info()
```

```python
#basic statics of the dataset
data.describe()
```

```python
import seaborn as sns
sns.heatmap(X.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
sns.pairplot(data, hue='Target')
plt.show()
```

```python
#plotting the graph before applying svd
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Plot before applying SVD')
plt.figure(figsize=(10,10))
plt.show()
```

```python
#applying truncatedSVD
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=2)
iris_transformed = svd.fit_transform(iris.data)
```

```python
#plotting the graph after applying svd
plt.scatter(iris_transformed[:, 0], iris_transformed[:, 1], c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Plot after applying SVD')
plt.show()
```

###<FONT COLOR = "GREEN "> Application of SVD for Image compression </FONT>

Image compression is a way to make digital images smaller in size without losing important information. One way to compress images is to use something called Singular Value Decomposition (SVD). This method breaks down an image into three smaller matrices, called U, S, and V^T.

The matrix S is the most important part, and it contains information about the image. By keeping only the largest values in S, we can compress the image while still keeping the important parts. For example, if we have a 1000x1000-pixel image, we can compress it from 1 MB to about 25 KB by keeping only the top 100 values in S.

To use SVD for image compression, we can use a programming language called Python and a library called numpy. We can load the image as a numpy array, convert it to grayscale (if it's not already), and apply SVD to the image matrix. Then, we can keep only the top values in S and use the other two matrices to reconstruct the compressed image.

This method is useful for things like medical images and satellite images, where we need to store or send lots of images but don't have a lot of space or bandwidth. Overall, SVD-based image compression is a powerful tool that can help us keep important information while making digital images much smaller.
